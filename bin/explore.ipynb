```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import re
import string
```

```python
data_path = '/home/michaelneuder/github/shakespeare/data'
```

```python
data = pd.read_csv(os.path.join(data_path, 'Shakespeare_data.csv'))
```

```python
data_richard = data[data.Play == "Richard III"]
```

```python
words = ''
remove = string.punctuation
pattern = r"[{}]".format(remove) 

for index, row in data.iterrows():
    words += re.sub(pattern, "",row['PlayerLine'].lower())
```

```python
words = words.split()
```

```python
counts = dict()
for word in words:
    counts[word] = counts.get(word, 0) + 1
```

```python
sorted_by_value = sorted(counts.items(), key=lambda kv: kv[1])
```

```python
sorted_max = sorted_by_value[::-1]
```
