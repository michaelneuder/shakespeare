```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import re
import scipy.stats as stats
import string
```

```python
np.random.uniform()
```

```python
x_vals = []
y_vals = []
for ii in range(30):
    xval = 2+2*np.random.uniform()
    yval = xval + 0.3*np.random.randn()
    if yval > 4:
        yval = 4
    x_vals.append(xval)
    y_vals.append(yval)
```

```python
plt.style.use('ggplot')
```

```python
f, ax = plt.subplots(figsize=(5,5))
ax.set_xlabel('high school GPA')
ax.set_ylabel('college school GPA')
ax.scatter(x_vals, y_vals)
plt.savefig('lin_reg.png')
```

```python
sl, i , r, p, std_er = stats.linregress(x_vals, y_vals)
```
